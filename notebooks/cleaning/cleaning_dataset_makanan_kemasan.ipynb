## Cleaning Data

```python
df.drop_duplicates(inplace=True)

duplicates_after_removal = df.duplicated().sum()
print(f"Jumlah baris duplikat setelah penghapusan: {duplicates_after_removal}")
```

```python
columns_to_fill = ['shelf_life_unopened_days', 'ideal_storage_temp', 'temp_coefficient', 'ph_level']

for col in columns_to_fill:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)

print("Missing values after median imputation:")
print(df[columns_to_fill].isnull().sum())
```

```python
df['food_category'] = df['food_category'].fillna('Unknown')
df['packaging_type'] = df['packaging_type'].fillna('Unknown')
df['shelf_life_opened_days'] = df['shelf_life_opened_days'].fillna('Unknown')

print("Missing values after imputation for categorical columns:")
print(df[['food_category', 'shelf_life_opened_days', 'packaging_type']].isnull().sum())
```

```python
df['food_category'] = df['food_category'].astype(str).str.lower().str.strip()
print("Unique values for 'food_category' after normalization:")
print(df['food_category'].unique())
```

```python
category_mapping = {
    'caned': 'kaleng',
    'snaks': 'camilan',
    'snack': 'camilan',
    'daory': 'produk susu',
    'beverag': 'minuman',
    'dairi': 'produk susu',
    'dairy': 'produk susu',
    'sauced': 'saus',
    'sauce': 'saus',
    'beverage': 'minuman',
    'canned': 'kaleng'
}
df['food_category'] = df['food_category'].replace(category_mapping)

print("Unique values for 'food_category' after correcting inconsistencies:")
print(df['food_category'].unique())
```

```python
df['packaging_type'] = df['packaging_type'].astype(str).str.lower().str.strip()
print("Unique values for 'packaging_type' after normalization:")
print(df['packaging_type'].unique())
```

```python
packaging_mapping = {
    'plastic': 'plastik',
    'glass bottle': 'botol kaca',
    'botol': 'botol kaca',
    'tin': 'kaleng'
}
df['packaging_type'] = df['packaging_type'].replace(packaging_mapping)

print("Unique values for 'packaging_type' after correcting inconsistencies:")
print(df['packaging_type'].unique())
```

```python
median_preservatives_level = df[df['preservatives_level'] >= 0]['preservatives_level'].median()
df.loc[df['preservatives_level'] < 0, 'preservatives_level'] = median_preservatives_level

print(f"Median for non-negative 'preservatives_level': {median_preservatives_level}")
print("Negative values in 'preservatives_level' after correction:")
print(df[df['preservatives_level'] < 0]['preservatives_level'].count())
```

```python
median_ph_level = df[df['ph_level'] >= 0]['ph_level'].median()
df.loc[df['ph_level'] < 0, 'ph_level'] = median_ph_level

print(f"Median for non-negative 'ph_level': {median_ph_level}")
print("Negative values in 'ph_level' after correction:")
print(df[df['ph_level'] < 0]['ph_level'].count())
```

```python
cleaned_shelf_life = df['shelf_life_opened_days'].copy()
cleaned_shelf_life = cleaned_shelf_life.astype(str).str.replace(' hari', '', regex=False).str.replace('hari', '', regex=False)
cleaned_shelf_life = cleaned_shelf_life.replace('Unknown', pd.NA)
df['shelf_life_opened_days'] = pd.to_numeric(cleaned_shelf_life, errors='coerce')

print("First 5 rows of df['shelf_life_opened_days'] after cleaning:")
display(df[['food_name', 'shelf_life_opened_days']].head())
```

```python
median_shelf_life_opened_days = df[df['shelf_life_opened_days'] >= 0]['shelf_life_opened_days'].median()
df.loc[df['shelf_life_opened_days'] < 0, 'shelf_life_opened_days'] = median_shelf_life_opened_days

print(f"Median for non-negative 'shelf_life_opened_days': {median_shelf_life_opened_days}")
print("Negative values in 'shelf_life_opened_days' after correction:")
print(df[df['shelf_life_opened_days'] < 0]['shelf_life_opened_days'].count())
```
