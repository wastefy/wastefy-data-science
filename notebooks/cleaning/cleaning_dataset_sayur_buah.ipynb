# Cleaning

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

## ethylene_producer

```python
copy_ethylene_producer = df['ethylene_producer'].astype(str).str.lower()
copy_ethylene_producer.unique()
```

```python
copy_ethylene_producer.value_counts()
```

```python
ethylene_producer_map = {
    'false': False,
    'true': True,
    '1': True,
    '0': False,
    'tidak': False,
    'iya': True,
    'no': False,
    'nan': False
}

copy_ethylene_producer = copy_ethylene_producer.map(ethylene_producer_map)
copy_ethylene_producer.value_counts()
```

```python
copy_ethylene_producer.fillna(False, inplace=True)
copy_ethylene_producer = copy_ethylene_producer.astype(bool)
copy_ethylene_producer.isnull().sum()
```

```python
df['ethylene_producer'] = copy_ethylene_producer
```

## freshness_score_visual

```python
copy_freshness_score_visual = df['freshness_score_visual'].astype(str).str.lower()
copy_freshness_score_visual.unique()
```

```python
copy_freshness_score_visual = pd.to_numeric(copy_freshness_score_visual, errors='coerce')
copy_freshness_score_visual.unique()
```

```python
copy_freshness_score_visual = copy_freshness_score_visual.fillna(copy_freshness_score_visual.mean())
copy_freshness_score_visual.unique()
```

```python
df['freshness_score_visual'] = copy_freshness_score_visual
```

## item_type

```python
copy_item_type = df['item_type']
copy_item_type.isnull().sum()
```

```python
copy_item_type = copy_item_type.fillna('Unknown')
copy_item_type.isnull().sum()
```

```python
df['item_type'] = copy_item_type
```

## respiration_rate

```python
copy_respiration_rate = df['respiration_rate'].astype(str).str.lower()
copy_respiration_rate.unique()
```

```python
copy_respiration_rate_map = {
    'l': 'rendah',
    'hi': 'tinggi',
    'med': 'sedang',
    'high ': 'tinggi',
    'medium': 'sedang',
    'high': 'tinggi',
    'low': 'rendah'
}

copy_respiration_rate = copy_respiration_rate.map(copy_respiration_rate_map)
```

```python
copy_respiration_rate.value_counts()
```

```python
copy_respiration_rate.isnull().sum()
```

```python
copy_respiration_rate = copy_respiration_rate.fillna('rendah')
copy_respiration_rate.isnull().sum()
```

```python
df['respiration_rate'] = copy_respiration_rate
```

## sensitivity_to_chilling

```python
copy_sensitivity_to_chilling = df['sensitivity_to_chilling'].astype(str).str.lower()
copy_sensitivity_to_chilling.unique()
```

```python
copy_sensitivity_to_chilling_map = {
    'true': True,
    'false': False,
    '1': True,
    '0': False,
    'no': False,
    'yes': True,
    'ya': True,
    'tidak': False
}

copy_sensitivity_to_chilling = copy_sensitivity_to_chilling.map(copy_sensitivity_to_chilling_map)
```

```python
copy_sensitivity_to_chilling.value_counts()
```

```python
copy_sensitivity_to_chilling.isnull().sum()
```

```python
mapping = {'High': True, 'Medium': False, 'Low': False}

copy_sensitivity_to_chilling = copy_sensitivity_to_chilling.fillna(df['respiration_rate'].map(mapping))

copy_sensitivity_to_chilling = copy_sensitivity_to_chilling.fillna(False).astype(bool)
copy_sensitivity_to_chilling.isnull().sum()
```

```python
df['sensitivity_to_chilling'] = copy_sensitivity_to_chilling
```

## standard_shelf_life_fridge

```python
copy_standard_shelf_life_fridge = df['standard_shelf_life_fridge']
copy_standard_shelf_life_fridge.unique()
```

```python
copy_standard_shelf_life_fridge = copy_standard_shelf_life_fridge.abs()
copy_standard_shelf_life_fridge.unique()
```

```python
copy_standard_shelf_life_fridge.isnull().sum()
```

```python
copy_standard_shelf_life_fridge = copy_standard_shelf_life_fridge.fillna(copy_standard_shelf_life_fridge.mean())
copy_standard_shelf_life_fridge.isnull().sum()
```

```python
df['standard_shelf_life_fridge'] = copy_standard_shelf_life_fridge
```

## standard_shelf_life_room

```python
copy_standard_shelf_life_room = df['standard_shelf_life_room']
copy_standard_shelf_life_room.unique()
```

```python
copy_standard_shelf_life_room.isnull().sum()
```

```python
copy_standard_shelf_life_room = copy_standard_shelf_life_room.fillna(copy_standard_shelf_life_room.mean())
copy_standard_shelf_life_room.isnull().sum()
```

```python
df['standard_shelf_life_room'] = copy_standard_shelf_life_room
```

## water_content_pct

```python
copy_water_content_pct = df['water_content_pct']
copy_water_content_pct.unique()
```

```python
copy_water_content_pct = copy_water_content_pct.replace('%', '', regex=True)
copy_water_content_pct.unique()
```

```python
copy_water_content_pct.isnull().sum()
```

```python
copy_water_content_pct = copy_water_content_pct.astype(float)
copy_water_content_pct = copy_water_content_pct.fillna(copy_water_content_pct.mean())
copy_water_content_pct.isnull().sum()
```

```python
df['water_content_pct'] = copy_water_content_pct
```
